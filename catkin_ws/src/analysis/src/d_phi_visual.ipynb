```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
i = 0
file_1 = "result/ssd_square_cw.txt"
file_2 = "result/ssd_square_ccw.txt"
file = open(file_1, 'r') 
start = 0
end = 100
all_text = file.readlines()
dis = []
phi = []
counter = 0
t = []
for text in all_text:
    counter += 1
    if counter >=end:
        continue
    text = text.replace('\n', '')
    split = text.split(' ')
    dis.append(float(split[0]))
    phi.append(float(split[1]))
    t.append(counter)

    file.close()
print(counter)
counter = 0
if i==0:  
    print("IN")
    file = open(file_2, 'r') 
    all_text = file.readlines()
    for text in all_text:
        
        if counter >=end:
            continue
        counter += 1
        text = text.replace('\n', '')
        split = text.split(' ')
        dis.append(float(split[0]))
        phi.append(float(split[1]))
        t.append(counter)
print(counter)
print("phi mean = ", round(np.mean(phi), 3))
print("phi standard = ", round(np.std(phi), 3))
print("distance mean = ", round(np.mean(dis), 3))
print("distance standard = ", round(np.std(dis), 3))

file.close()
```

```python
file_name = "ssd_eight_shape"
file = open("result/" + file_name + ".txt", 'r') 
all_text = file.readlines()
dis = []
phi = []
counter = 0
t = []
start = 0
end = 1000
for text in all_text:
    counter += 1
    if counter >= end or counter < start:
        continue
    text = text.replace('\n', '')
    split = text.split(' ')
    dis.append(float(split[0]))
    phi.append(float(split[1]))
    t.append(counter)

print("phi mean = ", np.mean(phi))
print("phi standard = ", np.std(phi))
print("distance mean = ", np.mean(dis))
print("distance standard = ", np.std(dis))

file.close()
```

```python
i = 0
if i==0:
    file_name_ssd = "ssd_square_cw"
    file_name_moos = "moos_square_ccw"
    file_name_combine = "combine_square"
    save_name = "square"
    limit = 110
elif i==1:
    file_name_ssd = "ssd_circle_ccw"
    file_name_moos = "moos_circle_ccw"
    file_name_combine = "combine_circle_ccw"
    save_name="circle"
    limit = 110
else:
    file_name_ssd = "ssd_eight_shape"
    file_name_moos = "moos_eight_shape"
    file_name_combine = "combine_eight_shape"
    save_name="eight_shape"
    limit = 180



file = open("result/" + file_name_ssd + "_first_trial.txt", 'r') 
all_text = file.readlines()
dis_ssd = []
phi_ssd = []
counter = 0
t_ssd = []
for text in all_text:
    counter += 1
    text = text.replace('\n', '')
    split = text.split(' ')
    dis_ssd.append(float(split[0]))
    phi_ssd.append(float(split[1]))
    t_ssd.append(counter)

file.close()


file = open("result/" + file_name_moos + "_first_trial.txt", 'r') 
all_text = file.readlines()
dis_moos = []
phi_moos = []
counter = 0
t_moos = []
for text in all_text:
    counter += 1
    text = text.replace('\n', '')
    split = text.split(' ')
    dis_moos.append(float(split[0]))
    phi_moos.append(float(split[1]))
    t_moos.append(counter)

file.close()

file = open("result/" + file_name_combine + "_first_trial.txt", 'r') 
all_text = file.readlines()
dis_combine = []
phi_combine = []
counter = 0
t_combine = []
for text in all_text:
    counter += 1
    text = text.replace('\n', '')
    split = text.split(' ')
    dis_combine.append(float(split[0]))
    phi_combine.append(float(split[1]))
    t_combine.append(counter)

file.close()
```

```python
plt.figure(figsize=(20,10))
plt.rcParams['axes.unicode_minus']=False
tick_size = 17
label_size = 25
legend_size = 20
plt.rcParams['xtick.labelsize']=tick_size
plt.rcParams['ytick.labelsize']=tick_size
plt.rcParams['xtick.major.pad']=18
plt.rcParams['ytick.major.pad']=18



plt.subplot(2, 1, 1)

plt.plot(t_ssd, phi_ssd, c='r', label='SSD')
plt.plot(t_moos, phi_moos, c='g', label='MOOS-IvP')
plt.plot(t_combine, phi_combine, c='b', label='Combine')

plt.ylabel("Delta Phi(rad)", fontsize=label_size)
y_tick = np.arange(-3, 4, 1.0)
x_tick = np.arange(0, limit, 5)
plt.ylim(-3, 4)
plt.xlim(0, limit)
plt.xticks(x_tick)
plt.yticks(y_tick)
plt.grid()
plt.legend(loc='upper right', fontsize=legend_size)

plt.subplot(2, 1, 2)

plt.plot(t_ssd, dis_ssd, c='r', label='SSD')
plt.plot(t_moos, dis_moos, c='g', label='MOOS-IvP')
plt.plot(t_combine, dis_combine, c='b', label='Combine')

plt.xlabel("Time(sec)", fontsize=label_size)
plt.ylabel("Delta Distance(meter)", fontsize=label_size)
y_tick = np.arange(0, 18, 3.0)

x_tick = np.arange(0, limit, 5)

plt.ylim(0, 15)
plt.xlim(0, limit)
plt.xticks(x_tick)
plt.yticks(y_tick)
plt.grid()
plt.gcf().subplots_adjust(bottom=0.15)
plt.legend(loc='upper right', fontsize=legend_size)

plt.savefig(save_name + '.png')

plt.show()
```
