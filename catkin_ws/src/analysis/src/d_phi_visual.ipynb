```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open("combine_cirlce_ccw.txt", 'r') 
all_text = file.readlines()
dis = []
phi = []
counter = 0
for text in all_text:
    counter += 1
    #if counter >= 180:
    #    break
    text = text.replace('\n', '')
    split = text.split(' ')
    dis.append(float(split[0]))
    phi.append(float(split[1]))

#print (phi)

print(np.mean(phi))
```

```python
plt.figure(figsize=(10,8))
plt.rcParams['axes.unicode_minus']=False

plt.scatter(phi, dis)
plt.xlabel("Phi", fontsize=18)
plt.ylabel("Distance", fontsize=18)

y_limit = 25
plt.xlim(-4, 4)
plt.ylim(0, y_limit)
x_tick = np.arange(-4, 4, 1.0)
y_tick = np.arange(0, y_limit, 1.0)
plt.xticks(x_tick)
plt.yticks(y_tick)

plt.grid()
plt.savefig('combine_cirlce_ccw.png')
plt.show()
```
